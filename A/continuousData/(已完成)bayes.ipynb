```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split as TTS

data = pd.read_csv(
    r"C:\Users\96361\iCloudDrive\数据挖掘实践-胡仕成\A\continuousData\Sampledata-continuous.CSV", encoding='GBK')


print(data)
```

```python
X_train = data.iloc[:, :7].to_numpy()
y_train = data.iloc[:, 7].to_numpy()
Xtrain, Xtest, Ytrain, Ytest = TTS(
    X_train, y_train, test_size=0.3, random_state=420)

print(X_train,type(X_train))
```

```python
print(Xtest,type(Xtest))
```

```python
Xtest_extra = [[0.12,	0, 0.72,	0.01, 0.02,	0.02,	0]]
Ytest_extra = [['2G']]

clf = GaussianNB()
# 拟合数据
clf.fit(Xtrain, Ytrain)
```

```python

print("==========================================")
print("测试准确率为:")
print(clf.score(Xtest, Ytest))
print("==========================================")

print("==========================================")
print("测试数据[0.12,0, 0.72,0.01,0.02,0.02,0]")
print("分类结果CUSTOMER_TYPE:"+str(clf.predict([[0.12,0, 0.72,0.01,0.02,0.02,0]])))
print("==========================================")
```
